# 退火算法求最小值


### $ 根据热力学的原理，在温度为T时，出现能量差为dE的降温的概率为P(dE)，表示为：$
# $ P(dE) = exp( dE/(k*T) ) $
### $ 其中k是一个常数，exp表示自然指数，且dE<0(温度总是降低的)。这条公式指明了 $

# parameter 参数内参数详解
## T:系统温度，初时应在高温
## T_min:收敛的最小温度
## r:降温的速率
## expr:最初的假设解
## jump_max:最大的冷区轮数

```python
def f():  # 待优化最小函数
    for x in statistics_module.sigmoid(p):
        yield x
expr, path, funcpath = simulate_anneal(f(),parameter={
                                               "T": 1,
                                               "T_min": 0,
                                               "r": 0.11,
                                               "expr": 0,
                                               "jump_max": 1000
                                           })
```
